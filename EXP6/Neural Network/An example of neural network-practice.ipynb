```python
# Generate toy data for binary classification
np.random.seed(123)

x_train = np.random.rand(200, 2)
y_train = np.random.randint(2, size=200)

x_test = np.random.rand(50, 2)
y_test = np.random.randint(2, size=50)
```

# An example of neural network using numpy

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def der_sigmoid(x):
    return x*(1-x)
```

```python
def softmax(x):
    return(np.exp(x)/np.exp(x).sum())
```

```python

one = np.ones(len(x_train))
x_train1 = np.hstack((x_train, one.reshape(len(x_train),1)))
y_train1 = to_categorical(y_train)


one = np.ones(len(x_test))
x_test1 = np.hstack((x_test, one.reshape(len(x_test),1)))
y_test1 = to_categorical(y_test)
```

```python
y_test
```

```python
iter_loss = []

inputsize = 2
outputsize = 2
lr = 0.0001 #learning rate
hn = 128    #number of neurons in the hidden layer

iter_n = 500

#randomly initilize the weihghts
W_1 = np.random.randn(2+1,hn) # first layer weight
W_2 = np.random.randn(hn+1,2) # second layer weight

# training iteration
for iter in range(iter_n):
    
    loss = 0
    acc = 0
    #for each sample
    for i, x in enumerate(x_train1):
        
        #forward computation
        z = np.dot(x, W_1)
        h1 = sigmoid(z)
        
        one2 = np.ones(1)
        h1_one = np.append(h1,one2)
        predicted = softmax(np.dot(h1_one,W_2))
        
        #accuracy
        p = np.argmax(predicted)
        yhat = np.argmax(y_train1[i])
        if p == yhat:
            acc += 1
        
        
        #backpropagation
        loss += 1/2*sum((predicted - y_train1[i])**2)
        
        delta2 = (predicted - y_train1[i]).reshape(1,outputsize)
        W2_df = np.dot(h1_one.reshape(hn+1,1), delta2)
        
        delta1 = np.dot(delta2, W_2.T)[0][0:hn]
        der = (delta1 * der_sigmoid(h1))
        
#         print(x.shape)
#         print(der.shape)
        W1_df = np.dot(x.reshape(inputsize+1,1),der.reshape(hn,1).T)
        
        #update with SGD
        W_1 -= lr * W1_df
        W_2 -= lr * W2_df
    
    # Print the loss every 10 epochs
    if iter % 10 == 0:
        print(iter, "loss: " ,loss, "acc: ", acc/len(x_train))
    iter_loss.append(loss)
```

```python
#training curve
from matplotlib.pyplot import plot

plot(range(iter_n), iter_loss)
```
